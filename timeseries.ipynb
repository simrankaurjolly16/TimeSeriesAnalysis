```python
# Perform Dickey-fuller test to test the differenced series for stationality

for i in result_monthly.lp_dol:
    x = result_monthly.lp_dol.values
    result_new = adfuller(x)
    
    #print('\033[1m' + i + '\033[0m')
    print('ADF Statistic:', result_new[0])
    print('p-value:', result_new[1])
        
    
# Didn't pass the stationary test
```

```python
# Differencing the dataset

df_diff = result_monthly.lp_dol.diff(periods=5)

diff = df_diff.dropna()

for i in diff:
    x = diff.values
    result = adfuller(x)
    #print('\033[1m' + i + '\033[0m')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    
# Now passed
```

```python
# with the following function we can select highly correlated features

def correlation(corr_matrix, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```
